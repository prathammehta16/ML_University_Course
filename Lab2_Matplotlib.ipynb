```python
#Importing matplotlib
import matplotlib.pyplot as plt
```

```python
#Simple line chart
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y)
plt.show()
```

```python
#Correlation chart
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.scatter(x, y)
plt.title('Correlation Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4)
plt.title('Histogram')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Plotting of Multivariate data
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
z = [5, 4, 3, 2, 1]

scatter = plt.scatter(x, y, c=z, cmap='viridis')
plt.colorbar(scatter)
plt.title('Multivariate Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Plot Pi Chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels)
plt.title('Pie Chart')
plt.show()
```
